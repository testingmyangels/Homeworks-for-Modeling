```python
# Salary Net Calculator Program

# Step 1: Read the inputs with error handling
try:
    gross = float(input("Enter the gross salary: "))
    children = int(input("Enter the number of children: "))
except:
    print("Error: Please enter valid numbers!")
    exit()

# Step 2: Determine the income tax percentage based on gross salary
if gross < 1000:
    tax_rate = 10
elif gross < 2000:
    tax_rate = 12
elif gross < 4000:
    tax_rate = 14
else:
    tax_rate = 18

# Step 3: Determine the child tax cut based on gross salary
if gross < 2000:
    child_cut = 1  # 1% per child
else:
    child_cut = 0.5  # 0.5% per child

# Step 4: Calculate total tax cut from children
total_child_cut = children * child_cut

# Step 5: Calculate the final tax rate (subtract child cuts from income tax)
final_tax_rate = tax_rate - total_child_cut

# Make sure tax rate doesn't go below 0
if final_tax_rate < 0:
    final_tax_rate = 0

# Step 6: Calculate the tax amount
tax_amount = gross * (final_tax_rate / 100)

# Step 7: Calculate the net salary
net = gross - tax_amount

# Step 8: Print the results
print("\n--- Salary Calculation ---")
print(f"Gross Salary: €{gross:.2f}")
print(f"Income Tax Rate: {tax_rate}%")
print(f"Child Tax Cut: {total_child_cut}%")
print(f"Final Tax Rate: {final_tax_rate}%")
print(f"Tax Amount: €{tax_amount:.2f}")
print(f"Net Salary: €{net:.2f}")
```
